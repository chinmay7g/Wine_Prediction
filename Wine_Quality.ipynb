```python
# Field Descriptions

#Fixed acidity: most wine acids involved are fixed or nonvolatile (do not evaporate readily)
#Volatile acidity: amount of acetic acid in wine - can be unpleasant, vinegary taste if too high?
#Citric acid: found in small quantities, can add ‘freshness’ and flavor to wines
#Residual sugar: sugar remaining after fermentation stops, rare < 1 gram/liter, > 45 grams/liter are considered sweet
#Chlorides: amount of salt in the wine
#Free sulfur dioxide: the free form of SO2 - prevents microbial growth and the oxidation of wine
#Total sulfur dioxide: free + bound forms of S02; in low concentrations, mostly undetectable in wine, free SO2 over 50 ppm, evident in the nose and taste of wine
#Density: the density of water is close to that of water (approx 1) depending on the percent alcohol and sugar content
#pH: acidic on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale
#Sulphates: anadditive which can contribute to S02 levels, acts as an antimicrobial and antioxidant.
#Alcohol: the percent alcohol content of the wine
```

# Modelling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score

#extracting values

X = red_wine.loc[: , ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
                      'chlorides', 'free sulfur dioxide', 'total sulfur dioxide',
                      'density', 'pH', 'sulphates', 'alcohol']].values
y = red_wine.loc[: , ['quality']].values

#scaling values 
scaler  = StandardScaler()
Xt = scaler.fit_transform(X)

#splitting

X_train , X_test , y_train , y_test = train_test_split(Xt , y , test_size=0.2 , random_state=0)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X_train , y_train)

lin_reg_preds = lin_reg.predict(X_test)

lin_reg_results = mse(lin_reg_preds, y_test)

lin_r2 = r2_score(lin_reg_preds , y_test)

print(f'lin_reg_results:{lin_reg_results}')
print(f'RMSE for Linear Regression:{(lin_reg_results)**0.5}')
print(f'R2 score for Linear Regression:{lin_r2}')

import pandas as pd

feature_names = list(red_wine.drop('quality',axis=1))
print(feature_names)

```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree = np.random.choice([2,3]))

X_poly = poly_features.fit_transform(Xt)

lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly , y)

lin_reg_poly_preds = lin_reg_poly.predict(X_poly)

lin_reg_poly_res = mse(lin_reg_poly_preds , y)

poly_r2 = r2_score(lin_reg_poly_preds , y)

print(f'Result for polynomial regression:{lin_reg_poly_res}')
print(f'RMSE for Polynomial Regression:{(lin_reg_poly_res)*0.5}')
print(f'R2 score for Poly regression:{poly_r2}')
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor(max_depth=3)

tree.fit(X_train , y_train)

tree_preds = np.ceil(tree.predict(X_test))

tree_results = mse(np.ceil(tree_preds) , y_test)

r2_tree = r2_score(np.ceil(tree_preds) , y_test)

print(f'MSE for decision tree:{tree_results}')
print(f'RMSE for decision tree:{(tree_results)*0.5}')
print(f'R2 for decision tree:{r2_tree}')

tree.score(X_test , y_test)
```

# Random Forests

```python
from sklearn.ensemble import RandomForestRegressor

rf_tree = RandomForestRegressor()

rf_params = dict(max_depth = range(2,12,2) , n_estimators = range(5,20,5))

rf_grid = GridSearchCV(rf_tree , rf_params , cv=4 , n_jobs=-1 , verbose=True)

rf_grid.fit(X_train , y_train)
```

```python
print(f'Score for random forest trees:{rf_grid.score(X_test,y_test)}')

print(f'Best params:{rf_grid.best_params_}')
```

```python
import pandas as pd

feature_names = list(red_wine.drop('quality',axis=1))
imp_vals = rf_grid.best_estimator_.feature_importances_

feature_imp_df = pd.DataFrame({'Imp. values':imp_vals} , index = feature_names)
feature_imp_df.plot(kind='bar')
```
